```python
import os
from ultralytics import YOLO
from google.colab import drive

# === 1. 사전 준비 ===
# Google Drive 마운트 확인
if not os.path.exists('/content/drive'):
    print("구글 드라이브를 마운트합니다...")
    drive.mount('/content/drive')

# [수정 필요] 모델 경로 및 동영상 파일 경로 설정
MODEL_PATH = '/content/drive/MyDrive/yolo11s/models/yolov11s_2class_zerodce_edsr.pt' # 기존 모델 경로 유지
VIDEO_PATH = '/content/edsr_test_720.mp4'  # 처리할 동영상 파일 경로 (여기를 수정하세요!)

# 경로 존재 여부 확인
if not os.path.exists(MODEL_PATH):
    raise FileNotFoundError(f"모델 파일을 찾을 수 없습니다: {MODEL_PATH}")
if not os.path.exists(VIDEO_PATH):
    raise FileNotFoundError(f"동영상 파일을 찾을 수 없습니다: {VIDEO_PATH}")

# === 2. 모델 로드 및 동영상 추론 실행 ===
print(f"모델 로드 중: {MODEL_PATH}")
model = YOLO(MODEL_PATH)

print(f"'{VIDEO_PATH}' 영상에 대해 추론을 시작합니다...")

# 동영상 예측 실행
# source에 동영상 경로를 넣으면 자동으로 프레임 단위로 처리하여 결과 영상을 저장합니다.
results = model.predict(
    source=VIDEO_PATH,
    save=True,       # 결과를 동영상 파일로 저장
    conf=0.3,        # 탐지 신뢰도 임계값
    imgsz=1280,      # 추론 시 이미지 크기 (동영상 해상도에 맞춰 조절 가능, 예: 1920, 1280 등)
    verbose=True     # 진행 상황 출력
)

# === 3. 결과 확인 ===
save_dir = results[0].save_dir
print(f"\n추론 완료! 처리된 영상이 저장되었습니다.")
print(f"저장 경로: {save_dir}")
print("해당 경로에서 결과 .avi 또는 .mp4 파일을 확인하세요.")
```

```python
# .avi 파일 .mp4 파일 변환

import os

# 여기에 변환하고 싶은 avi 파일 경로를 넣으세요
input_file = "/content/edsr_test_720_pred.avi"

# 변환 코드 (ffmpeg 사용)
output_file = os.path.splitext(input_file)[0] + ".mp4"
os.system(f"ffmpeg -y -i '{input_file}' -vcodec libx264 '{output_file}'")

print(f"변환 완료! 저장된 파일: {output_file}")
```
